```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import math
#%matplotlib inline
```

```python
def dis(p1,p2):
    dis = math.sqrt((p1[0]-p2[0]) ** 2 + (p1[1]-p2[1]) ** 2 + (p1[2]-p2[2]) ** 2)
    return dis
```

```python
cube1 = np.asarray([[0, 0, 0], [3, 0, 0], [0, 3, 0], [0, 0, 3], [0, 3, 3], [3, 0, 3], [3, 3, 0], [3, 3, 3]])
cube2 = np.asarray([[0, 2, 2], [1, 4, 4], [2, 0, 3], [2, 3, 0], [3, 2, 5], [3, 5, 2], [4, 1, 1], [5, 3, 3]])
```

```python
size = 100
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(cube1[:,0],cube1[:,1],cube1[:,2],s=100,c='b')
ax.scatter(cube2[:,0],cube2[:,1],cube2[:,2],s=100,c='r')
print()
print(dis(cube2[0],cube2[1]))
```
