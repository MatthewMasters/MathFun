```python
import math
from tqdm import tqdm
```

```python
def getL(t):
    dl = 0
    largest_side = t/4
    if largest_side % 1 == 0:
        dl += 1
    for le in range(3,math.ceil(largest_side)+9):
        side = math.sqrt(t + le ** 2)
        if side % 1 == 0:
            dl += 1
            if dl > t:
                break
    return dl
        
def N(n):
    c = 0
    smallest_t = 8
    for t in range(smallest_t,1000000+smallest_t):
        L = getL(t)
        if L == n:
            print(t)
            c += 1
    return c
```
