```python
r = 0.5
d = 2 * r
cA = math.pi * (r ** 2)
L = (1 - cA) / 4
print(L)
```

```python
#First, intersection of line and circle arc evaluated (x & y)
#A_a is the AUC of circle arc on right side of selection
#A_b is triangle on left side of intersection
n = 2
m = 1 / n
b = (-1/2) - m * (-1/2)
sq = -4 * (b ** 2) + (m ** 2) + 1
num = +math.sqrt(sq) + (2 * b * m)
den = 2 * (m ** 2 + 1)
x = -(num / den)
y = m * x + b
A_b = (1/2) * ((0.5-abs(x)) * (0.5-abs(y)))
A_a = (abs(x) * 0.5) - ((-2 * math.sqrt(1 - 4 * x ** 2) * x - math.asin(2 * x)) / 8)
A_tot = A_a + A_b
percent = A_tot / L
print(percent)
```
