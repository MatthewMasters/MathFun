```python
import numpy as np
```

```python
fo = open('Problem11_data.txt', 'r')

lines = []
for line in fo:
    data = str(line).rstrip().split(' ')
    arr = np.asarray(data, dtype=np.int64)
    lines.append(arr)

arr = np.stack(lines)
print(arr)
```

```python
high = 0
for xi in range(20):
    for yi in range(20):
        # Down sequence
        print('Head: %d,%d' % (xi,yi))
        if xi + 3 < 20:
            p0, p1, p2, p3 = arr[xi,yi], arr[xi+1,yi], arr[xi+2,yi], arr[xi+3,yi]
            product_down = p0 * p1 * p2 * p3
            print('down %d * %d * %d * %d = %d' % (p0, p1, p2, p3, product_down))
        if yi + 3 < 20:
            p0, p1, p2, p3 = arr[xi,yi], arr[xi,yi+1], arr[xi,yi+2], arr[xi,yi+3]
            product_right = p0 * p1 * p2 * p3
            print('rght %d * %d * %d * %d = %d' % (p0, p1, p2, p3, product_right))
        if xi + 3 < 20 and yi + 3 < 20:
            p0, p1, p2, p3 = arr[xi,yi], arr[xi+1,yi+1], arr[xi+2,yi+2], arr[xi+3,yi+3]
            product_right_diag = p0 * p1 * p2 * p3
            print('diagr %d * %d * %d * %d = %d\n' % (p0, p1, p2, p3, product_right_diag))
        if xi + 3 < 20 and yi - 3 >= 0:
            p0, p1, p2, p3 = arr[xi,yi], arr[xi+1,yi-1], arr[xi+2,yi-2], arr[xi+3,yi-3]
            product_left_diag = p0 * p1 * p2 * p3
            print('diagl %d * %d * %d * %d = %d\n' % (p0, p1, p2, p3, product_left_diag))
        highest = max([product_down, product_right, product_right_diag, product_left_diag])
        if highest > high:
            high = highest
```

```python
print(high)
```
