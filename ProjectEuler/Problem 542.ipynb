```python
def isGeometricProgression2(C): #Any Length
    C.sort(reverse=True)
    r = -1
    for en,el in enumerate(C[1:]):
        if r == -1:
            r = float(el)/C[0]
        else:
            r_2 = (float(el)/C[0]) ** (1.0/(en+1))
            #print('Ci: %d\tC1: %d\tr: %f\tidx: %d' % (el, C[0], r_2, en+1))
            if not r_2 == r:
                return False
    return True
```

```python
def isGeometricProgression(C):
    C.sort(reverse=True)
    r = float(C[1])/C[0]
    if (int((C[0]*(r ** 2))*1000)/1000.0).is_integer():
        return True
    else:
        return False
```

```python
def extendProgression(C):
    C.sort(reverse=True)
    next_val = -1
    r = float(C[1])/C[0]
    solved = False
    while not solved:
        if next_val != -1:
            C.append(int(next_val))
        next_val = int((C[0] * (r ** (len(C))))*1000)/1000.0
        if not next_val.is_integer():
            solved = True
        if next_val < 1:
            solved = True
    return C
```

```python
def S(k):
    integers = [i for i in range(1,k+1)]
    sums = []
    count = 0
    for combo in itertools.combinations(integers, 2):
        combo = list(combo)
        if isGeometricProgression(combo):
            count += 1
            if count > 1500:
                break
            combo_e = extendProgression(combo)
            sums.append(sum(combo_e))
    max_sum = max(sums)
    #print(sums.index(max_sum))
    return max_sum
```

```python
def T(n):
    ki = 4
    total_sum = 0
    #first_ki = ((-1) ** n) * S(n)
    while ki <= n:
        if ki > 500:
            test_ki = ki + 10
            part = ((-1) ** ki) * S(ki)
            test_part = ((-1) ** test_ki) * S(test_ki)
            if part == test_part:
                ki += 10
                continue
        part = ((-1) ** ki) * S(ki)
        total_sum += part
        print('ki: %d\t\tsum: %d\t\tpart: %d' % (ki, total_sum, part))
        ki += 1
    return total_sum
```
